```python
import json
import glob

# 输入目录和输出 JSONL 文件路径
input_dir = "./"  # WikiExtractor 生成的 JSON 文件夹
output_file = "./zhwiki.jsonl"  # 合并输出的 JSONL 文件路径

# 打开输出文件
with open(output_file, 'w', encoding='utf-8') as outfile:
    # 遍历所有 JSON 文件
    for filename in glob.glob(f"{input_dir}/*/wiki_*"):
        print(f"正在处理文件: {filename}")
        with open(filename, 'r', encoding='utf-8') as infile:
            # 逐行读取 JSON 内容
            # 每条数据只包含两个字段：id 和 contents
            for line in infile:
                data = json.loads(line)
                new_data = {"id": data['id'], "title": data["title"], "contents": data['text']}
                json.dump(new_data, outfile, ensure_ascii=False)
                outfile.write('\n')
```
